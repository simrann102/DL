## Perceptron Learning Algortihm

```python
def ercal(w1,w2,b,gate):
    x1=[0,0,1,1]
    x2=[0,1,0,1]
    output=[]
    error=[]
    for i in range(0,4):
        value=(x1[i]*w1)+(x2[i]*w2)-b
        if value>=0:
            output.append(1)
        else:
            output.append(0)
        error1=gate[i]-output[i]
        if error1==0:
            error.append(0)
        else:
            error.append(error1)
    print('x1=',x1)
    print('x2=',x2)
    print('y= ',gate)
    print('ŷ= ',output)
    print('ER=',error)
    print('w1=',w1,'  w2=',w2,'  b=',b)
    print('____________________________')
    return error
def PLA(gate):
    c = 0
    x1=[0,0,1,1]
    x2=[0,1,0,1]
    b=0
    w1=0.2
    w2=0.2
    error=ercal(w1,w2,b,gate)
    while sum(list(map(abs, error)))!=0 and c<=20:
        for i in range(0,4):
            if error[i]>0:
                w1=w1+(x1[i]*0.1)
                w2=w2+(x2[i]*0.1)
                b-=0.1
                break
            elif error[i]<0:
                w1=w1-(x1[i]*0.1)
                w2=w2-(x2[i]*0.1)
                b+=0.1
                break
        error=ercal(w1,w2,b,gate)
        c += 1
```

### AND gate

```python
PLA([0,0,0,1])
```

### OR gate

```python
PLA([0,1,1,1])
```

### NAND gate

```python
PLA([1,1,1,0])
```

### XOR gate

```python
PLA([0,1,1,0])
```

There is no optimal value hence, the PLA quits after 20 iteration

## Gradient Descent Algorithm

```python
import numpy as np
var = np.poly1d([1, -2, 2])
print("Polynomial function, f(x):\n", var)
derivative = var.deriv()
print("Derivative, f(x)'=", derivative)
starting=100
descent=[]
while derivative(starting)>0.02:
    descent.append(starting)
    starting-=(0.1*derivative(starting))
```

```python
import matplotlib.pyplot as plt

y = var(descent)
x_vals = np.linspace(-100, 100, 100)
y_vals = var(x_vals)
plt.plot(x_vals, y_vals)
plt.scatter(descent, y, c='r')
plt.show()
```
